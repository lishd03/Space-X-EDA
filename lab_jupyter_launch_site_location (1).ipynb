```python

# For each launch site, add a Circle object based on its coordinate (Lat, Long) values. In addition, add Launch site name as a popup label

import folium
from folium.map import Marker
from folium.features import DivIcon
import pandas as pd

# Define launch site data
launch_sites_data = {
    'Launch Site': ['CCAFS LC-40', 'CCAFS SLC-40', 'KSC LC-39A', 'VAFB SLC-4E'],
    'Lat': [28.562302, 28.563197, 28.573255, 34.632834],
    'Long': [-80.577356, -80.576820, -80.646895, -120.610745]
}

# Create DataFrame from launch site data
launch_sites = pd.DataFrame(launch_sites_data)

# Initialize the map
site_map = folium.Map(location=[28.5, -80], zoom_start=5)

# Function to add circle and marker for each launch site
def add_launch_site_marker(site_map, launch_sites):
    for index, site in launch_sites.iterrows():
        # Add circle marker
        folium.Circle(
            location=[site['Lat'], site['Long']],
            radius=1000,  # You can adjust the radius as needed
            color='#000000',
            fill=True
        ).add_child(folium.Popup(site['Launch Site'])).add_to(site_map)
        
        # Add marker with label
        folium.map.Marker(
            [site['Lat'], site['Long']],
            icon=DivIcon(
                icon_size=(20, 20),
                icon_anchor=(0, 0),
                html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % site['Launch Site']
            )
        ).add_to(site_map)

# Call the function to add launch site markers to the map
add_launch_site_marker(site_map, launch_sites)

# Display the map
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
